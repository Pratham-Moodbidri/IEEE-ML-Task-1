```python
import pandas as pd 
path = "/content/drive/MyDrive/tvmarketing.csv"
df = pd.read_csv(path)
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for col in df.columns:
    plt.hist(df[col])
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
for col in df.columns:
    plt.boxplot(df[col])
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

```python
plt.scatter(df['TV'], df['Sales'])
plt.title("Scatter plot of TV vs. Sales")
plt.xlabel("TV")
plt.ylabel("Sales")
plt.show()
```
